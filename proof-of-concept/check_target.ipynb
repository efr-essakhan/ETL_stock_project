```python
import boto3
import pandas as pd
from io import StringIO, BytesIO
from datetime import datetime, timedelta
```

## Retrieving data from S3

```python
s3 = boto3.resource('s3') #Have to do this always initally
bucket = s3.Bucket('stocks-etl-project-essa') #Retrieve the bucket with name from the server.
```

```python
#Get all objects in S3 bucket starting from arg_date_dt
objects = [obj for obj in bucket.objects.all()]
```

```python
objects 
```

```python
csv_obj_init = bucket.Object(key=objects[0].key).get().get('Body').read().decode('utf-8') #Get an initial csv, just to set the columns of the df.
data = StringIO(csv_obj_init) #convert the CSV object into string format.
df_init_csv = pd.read_csv(data, delimiter=',') #create a panda's df out of the csv
```

```python
parq_obj_init = bucket.Object(key=objects[2].key).get().get('Body').read() #Get an initial csv, just to set the columns of the df.
data = BytesIO(parq_obj_init) #convert the parq object into string format.
df_init_parq = pd.read_parquet(data) #create a panda's df out of the csv
```
